```python
import requests
import json
import os


import matplotlib.pyplot as plt
import numpy as np
from collections import Counter


counter = Counter()

review_dir = r"./reviews"
for review_file in os.listdir(review_dir):
    review_path = os.path.join(review_dir, review_file)

    if not review_path.endswith(".json"):
        continue        

    with open(review_path, 'r', encoding="utf-8-sig") as f:
        data = json.load(f)
        for review in data:
            content_len = len(review.get('content'))
            if content_len not in counter.keys():
                counter[content_len] = 1
            else:
                counter[content_len] += 1

review_dir = r"./reviews_keyboard"
for review_file in os.listdir(review_dir):
    review_path = os.path.join(review_dir, review_file)

    if not review_path.endswith(".json"):
        continue        

    with open(review_path, 'r', encoding="utf-8-sig") as f:
        data = json.load(f)

        for review in data:
            content_len = len(review.get('content'))
            if content_len not in counter.keys():
                counter[content_len] = 1
            else:
                counter[content_len] += 1








    




```

```python
# 데이터 준비
content_lengths = list(counter.keys())
frequencies = list(counter.values())

# 최댓값과 최솟값 확인
max_length = max(content_lengths)
min_length = min(content_lengths)

# 구간 설정
bins = np.linspace(min_length, max_length, 51)  # 50개의 구간

# 히스토그램 그리기

plt.figure(figsize=(12, 6))
plt.hist(content_lengths, bins=bins, weights=frequencies, edgecolor='black')
plt.title('Histogram of Content Lengths')
plt.xlabel('Content Length')
plt.ylabel('Frequency')

# x축 눈금 설정
plt.xticks(bins)
plt.yscale('log')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```
